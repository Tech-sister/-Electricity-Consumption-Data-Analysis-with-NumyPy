```python
import numpy as np
```

Problem Statement

You are provided with electricity consumption data (in kWh) for a household measured daily over 4 months (120 days). The daily consumption values range between 200 kWh (minimum) and 600 kWh (maximum). Your task is to simulate the dataset using NumPy and perform an in-depth analysis of the household’s electricity consumption.


```python
#1.Generate the Dataset


#Use numpy.random.randint() to simulate 120 daily consumption values between 200 and 600 kWh.
#For reproducibility


np.random.seed(42)

# Generate 120 daily electricity consumption values
consumption = np.random.randint(200, 601, 120)


print("Electricity Consumption Data (First 20 Days):")
print(consumption[:20])
```

```python
#2. Basic Statistics
#Calculate the mean, median, minimum, maximum, and standard deviation of daily consumption.

mean_consumption = np.mean(consumption)
median_consumption = np.median(consumption)
min_consumption = np.min(consumption)
max_consumption = np.max(consumption)
std_consumption = np.std(consumption)

print("\n--- Basic Statistics ---")
print("Mean Consumption:", mean_consumption)
print("Median Consumption:", median_consumption)
print("Minimum Consumption:", min_consumption)
print("Maximum Consumption:", max_consumption)
print("Standard Deviation:", std_consumption)
```

```python
#3.Identify Extremes


#Find the day number(s) with the highest and lowest consumption.

highest_days = np.where(consumption == max_consumption)[0] + 1
lowest_days = np.where(consumption == min_consumption)[0] + 1

print("\n--- Extreme Consumption Days ---")
print("Highest Consumption Day(s):", highest_days)
print("Lowest Consumption Day(s):", lowest_days)

```

```python
#4.Monthly Splitting
#Divide the dataset into 4 months (30 days each).
#Store each month’s data in a separate NumPy array.

month1 = consumption[0:30]

month2 = consumption[30:60]

month3 = consumption[60:90]

month4 = consumption[90:120]
```

```python
#5.Monthly Analysis
#For each month: calculate the total consumption and average daily consumption.
months = [month1, month2, month3, month4]

print("\n--- Monthly Analysis ---")

for i, month in enumerate(months, start=1):
    total = np.sum(month)
    average = np.mean(month)

    print(f"\nMonth {i}")
    print("Total Consumption:", total)
    print("Average Daily Consumption:", average)
```

```python
#6.Weekly Analysis
#Reshape or split the data into weeks of 7 days (ignore the extra 4 days at the end).
#Compute the weekly totals and weekly averages.

weekly_data = consumption[:119]
weeks = weekly_data.reshape(17, 7)

weekly_totals = np.sum(weeks, axis=1)
weekly_averages = np.mean(weeks, axis=1)

print("\n--- Weekly Analysis ---")

for i in range(len(weekly_totals)):
    print(f"Week {i+1}")
    print("Total:", weekly_totals[i])
    print("Average:", weekly_averages[i])
```

```python
#7.Overall Consumption Trend

#Calculate the cumulative consumption across the 120 days.
#Display the first 10 values of the cumulative sum.

cumulative_consumption = np.cumsum(consumption)

print("\n--- Cumulative Consumption ---")
print("First 10 Cumulative Values:")
print(cumulative_consumption[:10])

```

```python
#8.Anomaly Detection
#Identify the day numbers that are anomalies.

upper_limit = mean_consumption + (2 * std_consumption)
lower_limit = mean_consumption - (2 * std_consumption)

anomalies = np.where(
    (consumption > upper_limit) |
    (consumption < lower_limit))[0] + 1

print("\n--- Anomaly Detection ---")
print("Upper Limit:", upper_limit)
print("Lower Limit:", lower_limit)
print("Anomalous Day(s):", anomalies)

#Define anomalies as days where consumption is 2 standard deviations above or below the mean.
above_average_days = np.sum(consumption > mean_consumption)
below_average_days = np.sum(consumption < mean_consumption)

```
